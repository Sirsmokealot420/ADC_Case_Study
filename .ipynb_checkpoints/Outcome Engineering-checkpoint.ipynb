```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import matplotlib
import os
import math

#os.chdir("C:/Users/phaim/OneDrive/Applied Data Science/Fallstudie")

data_lgd = pd.read_csv("LGD_Panel.csv")
```

Datensatz eingrenzen: für die Analyser sind lediglich Beobachtungen relevant, die bereits abgeschlossene Defaults beschreiben. Daher werden Beobachtungen mit 'default_time' == 0 und 'rest_time' == NA vom Datensatz entfernt.

```python
data_lgd = data_lgd[data_lgd['default_time'] == 1]
data_lgd = data_lgd.dropna(subset = ['res_time'])


data_lgd.head()
```

Um die potenielle Auswahl von Features zu erweitern, erstellen wir auf Basis schon besteheneder Variablen weitere Regressoren. Hierbei handelt es sich um Differenzen, Zeitvariablen und quadriatischer Modellierungen. Der spätere Algorithmus kann somit einem größerer Anteil der Variation im Outcome erklären.

```python
data_lgd.loc[:, "res_duration"] = data_lgd.loc[:, "res_time"] - data_lgd.loc[:, 'time']
data_lgd.loc[:, "planned_lifetime"] = data_lgd.loc[:, "mat_time"] - data_lgd.loc[:, "orig_time"]
data_lgd.loc[:, "realized_lifetime"] = data_lgd.loc[:, "time"] - data_lgd.loc[:, "orig_time"]
data_lgd.loc[:, "rel_realized_lifetime"] = data_lgd.loc[:, "realized_lifetime"] / data_lgd.loc[:, "planned_lifetime"]
data_lgd.loc[:, "remaining_lifetime"] = data_lgd.loc[:, "planned_lifetime"] - data_lgd.loc[:, "realized_lifetime"]
data_lgd.loc[:, "diff_LTV"] = data_lgd.loc[:, "LTV_time"] - data_lgd.loc[:, "LTV_orig_time"]
data_lgd.loc[:, "rel_diff_LTV"] = data_lgd.loc[:, "LTV_time"] / data_lgd.loc[:, "LTV_orig_time"]


data_lgd.loc[:, "sq_realized_lifetime"] = data_lgd.loc[:, "realized_lifetime"]**2
data_lgd.loc[:, "sq_planned_lifetime"] = data_lgd.loc[:, "planned_lifetime"]**2
data_lgd.loc[:, "sq_remaining_lifetime"] = data_lgd.loc[:, "remaining_lifetime"]**2
data_lgd.loc[:, "sq_diff_LTV"] = data_lgd.loc[:, "diff_LTV"]**2
data_lgd.loc[:, "sq_res_duration"] = data_lgd.loc[:, "res_duration"]**2
data_lgd.loc[:, "sq_interest_rate_time"] = data_lgd.loc[:, "interest_rate_time"]**2
data_lgd.loc[:, "sq_LTV_time"] = data_lgd.loc[:, "LTV_time"]**2
data_lgd.loc[:, "sq_LTV_orig_time"] = data_lgd.loc[:, "LTV_orig_time"]**2
data_lgd.loc[:, "sq_balance_orig_time"] = data_lgd.loc[:, "balance_orig_time"]**2
data_lgd.loc[:, "sq_FICO_orig_time"] = data_lgd.loc[:, "FICO_orig_time"]**2

data_lgd["log_res_duration"] = np.log2(data_lgd['res_duration'])
```

Da die Rückgewinnung von Cashflows nach dem Ausfalleregnis mehrere Perioden überdauert, werden diese Zahlungen mit dem risikolosen Zinsatz auf die Ausfallperioden diskontiert. Wir wählen den risikolosen Zinsatz, da durch die ex-post Perspektive Minderungen in den zurückerhaltenen Cashlfows ausgeschlossen sind. Im Zuge dessen wird die diskontierte LGD-Rate erstellt.

```python
data_lgd.loc[:, 'recovery_res'] = data_lgd.loc[:, 'recovery_res'].astype(float)
data_lgd.loc[:, 'DLGD'] = (data_lgd.loc[:, 'balance_time'] - data_lgd.loc[:, 'recovery_res'] / (1 + 
            data_lgd.loc[:, 'rate_time'] / (100*4)) ** data_lgd.loc[:, 'res_duration']) / data_lgd.loc[:, 
                                                                                        'balance_time']
data_lgd.head()

sns.distplot(data_lgd['DLGD'], color='grey')
plt.show()
```

Wir winsorizen die neu erstellte diskontierte LGD (DLGD). Ausprägungen werden maximal auf eins und minimal auf 0.0001 beschränkt. Somit wird die Analyse robuster gegenüber Ausreißer und wir behalten uns die Möglichkeit bei, eventuell eine geloggtes Outcome zu schätzen. 
Ausprägungen der DLGD kleiner Null entstehen durch das Diskontieren. Ausprägungen größer eins ergeben sich aus einem (fast) Komplettausfall der Kreditsumme, kombiniert mit zusätzlichen Kosten des Ausfallprozesses. So ein Fall tritt im Datensatz "-" mal auf, mit einer durchschnittlichen LGD von "_" (begrenzt auf DLGD > 1). Durch das Winsorizen könnte eventuell die Erwartungstreue des Modells verletzt werden. Doch die Verringerung der späteren Schätzvarianz und der Zugewinn an Robustheit gegenüber Ausreißer, erlaubt einen allgemein besseren Bias-Variance trade-off. 

```python
data_lgd.loc[data_lgd['DLGD'] > 1, 'DLGD'] = 1
data_lgd.loc[data_lgd['DLGD'] <= 0, 'DLGD'] = 0.0001


data_lgd.DLGD.hist(bins = 40)

data_lgd.DLGD.describe().round(decimals = 2)
plt.ylabel('Frequency')
sns.distplot(data_lgd['DLGD'], color='grey')
plt.show()
```

```python
data_lgd["log_DLGD"] = np.log2(data_lgd['DLGD'])

data_lgd.log_DLGD.hist(bins = 40)
```
